```python
# Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose, STL
from statsmodels.tsa.stattools import kpss
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from prophet import Prophet
from pmdarima import auto_arima
import warnings
warnings.filterwarnings("ignore")

# Load Dataset
# Assuming the file is uploaded to the Colab environment and is in the default /content/ directory
file_path = "Dataset.xlsx"
df = pd.read_excel(file_path, sheet_name=0)
df.columns = ['Date', 'Prod']
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df = df.asfreq('MS')

# Time Series Creation
volume_ts = df['Prod']

# Train-Test Split
train = volume_ts[:'2016-10']
test = volume_ts['2016-11':]

# Decomposition
additive_decomp = seasonal_decompose(train, model='additive')
additive_decomp.plot()
plt.suptitle("Additive Decomposition")
plt.show()

multiplicative_decomp = seasonal_decompose(train, model='multiplicative')
multiplicative_decomp.plot()
plt.suptitle("Multiplicative Decomposition")
plt.show()

stl = STL(train, seasonal=13)
stl_result = stl.fit()
stl_result.plot()
plt.suptitle("STL Decomposition")
plt.show()

# Naive, Mean, Drift Forecasts
naive_pred = pd.Series(train.iloc[-1], index=test.index)
mean_pred = pd.Series(train.mean(), index=test.index)
drift = (train.iloc[-1] - train.iloc[0]) / (len(train) - 1)
drift_pred = pd.Series([train.iloc[-1] + drift * (i+1) for i in range(len(test))], index=test.index)

# Holt-Winters Multiplicative Model
hw_model = ExponentialSmoothing(train, seasonal='mul', seasonal_periods=12).fit()
hw_forecast = hw_model.forecast(len(test))

# ARIMA
arima_model = auto_arima(train, seasonal=True, m=12, trace=True)
arima_forecast = pd.Series(arima_model.predict(n_periods=len(test)), index=test.index)

# Prophet Forecast
prophet_df = train.reset_index().rename(columns={'Date': 'ds', 'Prod': 'y'})
prophet_model = Prophet()
prophet_model.fit(prophet_df)
future = prophet_model.make_future_dataframe(periods=len(test), freq='MS')
forecast = prophet_model.predict(future)
prophet_forecast = forecast.set_index('ds')['yhat'][-len(test):]

# Linear Regression with Trend and Seasonality
train_df = train.reset_index()
train_df['time'] = np.arange(len(train_df))
train_df = pd.get_dummies(train_df, columns=['Date'], drop_first=True)
X = train_df.drop(['Prod'], axis=1)
y = train_df['Prod']
reg = LinearRegression().fit(X, y)

# Accuracy Function
def accuracy_metrics(actual, predicted):
    return {
        'MAE': mean_absolute_error(actual, predicted),
        'RMSE': np.sqrt(mean_squared_error(actual, predicted))
    }

# Evaluate Forecasts
print("Naive:", accuracy_metrics(test, naive_pred))
print("Mean:", accuracy_metrics(test, mean_pred))
print("Drift:", accuracy_metrics(test, drift_pred))
print("Holt-Winters:", accuracy_metrics(test, hw_forecast))
print("ARIMA:", accuracy_metrics(test, arima_forecast))
print("Prophet:", accuracy_metrics(test, prophet_forecast))

# Plot Forecasts
plt.figure(figsize=(12,6))
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.plot(naive_pred, label='Naive')
plt.plot(mean_pred, label='Mean')
plt.plot(drift_pred, label='Drift')
plt.plot(hw_forecast, label='Holt-Winters')
plt.plot(arima_forecast, label='ARIMA')
plt.plot(prophet_forecast, label='Prophet')
plt.title("Forecast Comparison")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose, STL
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import kpss
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from scipy.stats import boxcox
import warnings
warnings.filterwarnings('ignore')

# Read data
volume = pd.read_excel("Dataset.xlsx", sheet_name=0)
print(volume.head())

# Create time series
volume_ts = pd.Series(volume.iloc[:, 1].values,
                     index=pd.date_range('2006-04', periods=len(volume), freq='M'))

print(f"TS Info: {volume_ts.index[0]} to {volume_ts.index[-1]}, Length: {len(volume_ts)}")

# Plot
plt.figure(figsize=(10, 6))
plt.plot(volume_ts)
plt.title('Volume Time Series')
plt.show()

# Seasonal plots
seasonal_decompose(volume_ts, model='additive').plot()
plt.show()

# Train/Test split
volume_train = volume_ts[:'2016-10']
volume_test = volume_ts['2016-11':]

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(volume_train)
plt.title('Train')
plt.subplot(1, 2, 2)
plt.plot(volume_test)
plt.title('Test')
plt.show()

# Box-Cox
_, lambda_val = boxcox(volume_ts + 1e-6)
print(f"Box-Cox lambda: {lambda_val:.6f}")

# Decompositions
seasonal_decompose(volume_ts, model='additive').plot()
plt.suptitle('Additive')
plt.show()

seasonal_decompose(volume_ts, model='multiplicative').plot()
plt.suptitle('Multiplicative')
plt.show()

STL(volume_ts, seasonal=13).fit().plot()
plt.suptitle('STL')
plt.show()

# Simple forecasts
h = 55
avgforecast = np.full(h, volume_train.mean())
naiveforecast = np.full(h, volume_train.iloc[-1])
seasonalnaive = np.array([volume_train.iloc[-(12-i%12)] for i in range(h)])
driftforecast = volume_train.iloc[-1] + (volume_train.iloc[-1] - volume_train.iloc[0]) / (len(volume_train)-1) * np.arange(1, h+1)

# Accuracy function
def accuracy(forecast, actual):
    forecast = forecast[:len(actual)]
    return {
        'RMSE': np.sqrt(np.mean((actual - forecast)**2)),
        'MAE': np.mean(np.abs(actual - forecast)),
        'MAPE': np.mean(np.abs((actual - forecast) / actual)) * 100
    }

# Print accuracies
print("Accuracies:")
print("Mean:", accuracy(avgforecast, volume_test))
print("Naive:", accuracy(naiveforecast, volume_test))
print("Seasonal Naive:", accuracy(seasonalnaive, volume_test))
print("Drift:", accuracy(driftforecast, volume_test))

# Holt-Winters
hw_model = ExponentialSmoothing(volume_train, seasonal_periods=12, trend='add', seasonal='mul').fit()
hw_forecast = hw_model.forecast(h)
print("Holt-Winters:", accuracy(hw_forecast, volume_test))

plt.figure(figsize=(12, 6))
plt.plot(volume_train.index, volume_train, label='Train')
plt.plot(volume_test.index, volume_test, label='Actual')
plt.plot(volume_test.index, hw_forecast[:len(volume_test)], '--', label='HW Forecast')
plt.legend()
plt.show()

# KPSS test
kpss_stat, p_val, _, _ = kpss(volume_train)
print(f"KPSS: {kpss_stat:.6f}, p-value: {p_val:.6f}")

# Auto ARIMA (simplified)
best_aic = np.inf
best_model = None
for p in range(3):
    for d in range(2):
        for q in range(3):
            try:
                model = ARIMA(volume_train, order=(p,d,q), seasonal_order=(1,1,1,12)).fit()
                if model.aic < best_aic:
                    best_aic = model.aic
                    best_model = model
            except: pass

print(f"Best ARIMA AIC: {best_aic:.2f}")
print(best_model.summary())

arima_forecast = best_model.forecast(h)
print("ARIMA:", accuracy(arima_forecast, volume_test))

plt.figure(figsize=(12, 6))
plt.plot(volume_train.index, volume_train, label='Train')
plt.plot(volume_test.index, volume_test, label='Actual')
plt.plot(volume_test.index, arima_forecast[:len(volume_test)], '--', label='ARIMA')
plt.legend()
plt.show()

# Neural Network
def create_features(data, lags=[1,12]):
    X, y = [], []
    for i in range(max(lags), len(data)):
        X.append([data[i-lag] for lag in lags])
        y.append(data[i])
    return np.array(X), np.array(y)

X_train, y_train = create_features(volume_train.values)
scaler_X, scaler_y = StandardScaler(), StandardScaler()
X_scaled = scaler_X.fit_transform(X_train)
y_scaled = scaler_y.fit_transform(y_train.reshape(-1,1)).ravel()

nn_model = MLPRegressor(hidden_layer_sizes=(10,), max_iter=1000, random_state=42)
nn_model.fit(X_scaled, y_scaled)

# NN forecast
nn_forecast = []
data_ext = list(volume_train.values)
for i in range(h):
    features = [data_ext[-1], data_ext[-12]]
    pred = scaler_y.inverse_transform([[nn_model.predict(scaler_X.transform([features]))[0]]])[0][0]
    nn_forecast.append(pred)
    data_ext.append(pred)

print("Neural Network:", accuracy(nn_forecast, volume_test))

# Linear models
n = len(volume_train)
trend = np.arange(1, n+1)
months = pd.get_dummies([d.month for d in volume_train.index])

# Model 1: trend + season
X1 = np.column_stack([trend, months])
reg1 = LinearRegression().fit(X1, volume_train)

# Model 2: trend + trend^2 + season
X2 = np.column_stack([trend, trend**2, months])
reg2 = LinearRegression().fit(X2, volume_train)

# Test predictions
trend_test = np.arange(n+1, n+len(volume_test)+1)
months_test = pd.get_dummies([d.month for d in volume_test.index])

X1_test = np.column_stack([trend_test, months_test])
X2_test = np.column_stack([trend_test, trend_test**2, months_test])

reg1_forecast = reg1.predict(X1_test)
reg2_forecast = reg2.predict(X2_test)

print("Linear Reg 1:", accuracy(reg1_forecast, volume_test))
print("Linear Reg 2:", accuracy(reg2_forecast, volume_test))




plt.figure(figsize=(16, 10))
plt.plot(volume_train.index, volume_train, label='Training Data', color='black', linewidth=1)
plt.plot(volume_test.index, volume_test, label='Actual Test Data', color='blue', linewidth=2)

# Plot all forecasts
plt.plot(forecast_index[:len(volume_test)], hw_forecast[:len(volume_test)],
         label='Holt-Winters', linestyle='--', alpha=0.8)
plt.plot(forecast_index[:len(volume_test)], arima_forecast[:len(volume_test)],
         label='ARIMA', linestyle='--', alpha=0.8)
plt.plot(forecast_index[:len(volume_test)], nn_forecast[:len(volume_test)],
         label='Neural Network', linestyle='--', alpha=0.8)
plt.plot(forecast_index[:len(volume_test)], reg1_forecast,
         label='Linear Reg (trend+season)', linestyle='--', alpha=0.8)
plt.plot(forecast_index[:len(volume_test)], reg2_forecast,
         label='Linear Reg (trend^2+season)', linestyle='--', alpha=0.8)

plt.title('Comparison of All Forecasting Methods')
plt.xlabel('Year')
plt.ylabel('Production Volume')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Summary of all accuracy metrics
print("\n" + "="*60)
print("SUMMARY OF ALL FORECASTING METHODS")
print("="*60)

methods = {
    'Mean Forecast': mean_acc,
    'Naive Forecast': naive_acc,
    'Seasonal Naive': snaive_acc,
    'Drift Forecast': drift_acc,
    'Holt-Winters': hw_acc,
    'ARIMA': arima_acc,
    'Neural Network': nn_acc,
    'Linear Reg (trend+season)': reg1_acc,
    'Linear Reg (trend^2+season)': reg2_acc
}

for method, metrics in methods.items():
    print(f"\n{method}:")
    for key, value in metrics.items():
        print(f"  {key}: {value:.4f}")

print("\n" + "="*60)

```
